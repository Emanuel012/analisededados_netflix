```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2020", periods=1000))

ts = ts.cumsum()

ts.plot();
```

```python
plt.show();
```

```python
df = pd.DataFrame(
    np.random.randn(1000, 4), index=ts.index, columns=["rating", "stars", "votes", "duration"]
)
```

```python
df = df.cumsum()

plt.figure();

df.plot();

plt.legend(loc='best');
```

```python
#As avaliações variam de acordo com os meses de férias(janeiro , julho) , acompanhado pela duração dos shows. Os votos tendem a aumentar de forma 
#sazonal. O número de pessoas que avalia o show diminuiu nos últimos dois anos.
```
